## Regression Discontinuity
Use RD when we don't have random assignment, but where some change occurs due to our 'running variable'.

MLDA study - Maximum Legal Drinking Age study

$D_a$ = 1 if age >= 21 else 0

Term | Definition
-|-
Running Variable | variable which determines treatment (age in this case)
Sharp RD | Treatment switches on and off immediately (e.g. in this case, at the 21st birthday)
Fuzzy RD | Treatment becomes stronger or more likely at some cutoff


### Regression setup:
$$\bar{M_a} = \alpha + \rho D_a + \gamma a + e_a$$

In example:

$\bar{M_a}$ : Death rate in month a


```python
%matplotlib inline
import numpy as np
np.random.seed(100)

#Import
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

#Create some death rate data:
x_axis = np.arange(0, 200)

cutoff = round(len(x_axis) / 2)

#Create y-axis data
#First half of data:   20*U - 0.01x
#Second half of data:  10*U - 0.01x
y_axis = list(map(lambda x : int(abs(np.random.normal(40 + 10*(x>cutoff), 8)) - 0.01 * x), x_axis))
```

```python
plt.plot(x_axis, y_axis)
plt.xlabel("Months Before/After Age 21")
plt.ylabel("Monthly Deaths")
```

## Regression Discontinuity:
$$\bar{M_a} = \alpha + \rho D_a + \gamma a + e_a$$

```python
#Set up variables
M_a = y_axis
D_a = x_axis >= cutoff
a = x_axis

#Massage data into DF:
in_df = pd.DataFrame(data = {"D_a" : D_a,
                              "a" : a,
                              "M_a" : M_a},
                    index = a)
```

```python
#Create Regression Objects
from sklearn.linear_model import LinearRegression
lr = LinearRegression()

#Fit the regressions:
lr.fit(X = in_df[["D_a", "a"]], y = in_df[["M_a"]])

#Find the difference at the intercept:
rho = lr.coef_[0][0]
print(lr.coef_[0])
```

```python
#Create simple regression function
def reg_f(x1, x2, lr):
    return(lr.intercept_[0] + x1 * lr.coef_[0][0] + x2 * lr.coef_[0][1])
```

```python
#Plot the regressions:
#Base data
plt.plot(x_axis, y_axis)
plt.xlabel("Months Before/After Age 21")
plt.ylabel("Monthly Deaths")

#Regressions:
start = x_axis[0]
end = x_axis[-1]
#PRE
plt.plot([start, cutoff], [reg_f(0, start, lr), reg_f(0, cutoff, lr)])
#POST
plt.plot([cutoff, end], [reg_f(1, cutoff, lr), reg_f(1, end, lr)])
```

### Summary stats:
#### Standard Error of sample mean:

```python
import scipy as sp
print("standard_error:", sp.std(y_axis) / sp.sqrt(len(y_axis)))
```

This is much smaller than our discontinuity coefficient and thus we can be confident that the change is significant.

### Limitations:
One limitation is that we have to assume the relationship is linear (or polynomial if we are fitting a poly regression). If our assumption is wrong, we will misjudge the discontinuity (which may even be as dramatic as seeing a discontinuity where there is really only non-linearity. Thus we need to concern ourselves with two cases;
1. There is discontinuity AND non-linearity
2. There is only non-linearity

There are two strategies here.  
First, we can try to model the non-linearity. One way is to add a quadratic control for the running variable.
$$\bar{M_a} = \alpha + \rho D_a + \gamma_1 a + \gamma_2 a^2 + e_a$$

```python
#Polynomial case:

#Create data:
x_axis = np.arange(0, 200)
cutoff = round(len(x_axis) / 2)
y_axis = list(map(lambda x : int(abs(np.random.normal(40 + 10*(x>cutoff), 8)) - 0.01 * x), x_axis))

#Massage data into nice structure:
M_a = y_axis
D_a = x_axis >= cutoff
#(a - a_0):
a = x_axis - cutoff
#delta:
a_squared = a ** 2

#Massage data into DF:
in_df = pd.DataFrame(data = {"D_a" : D_a, 
                             "a" : a, 
                             "a_squared" : a_squared,
                             "M_a" : M_a}, index = a)

#Run regression
lr = LinearRegression()
lr.fit(X = in_df[["D_a", "a", "a_squared"]], y = in_df[["M_a"]])

plt.plot(x_axis - cutoff, y_axis)
plt.xlabel("Months Before/After Age 21")
plt.ylabel("Monthly Deaths")

#Function for getting result:
def reg_f(x1, x2, x3, lr):
    return(lr.intercept_[0] + x1 * lr.coef_[0][0] + x2 * lr.coef_[0][1] + x3 * lr.coef_[0][2])

#Plot
plt.plot(a[:cutoff], 
         [x[0] for x in lr.predict(in_df.ix[in_df["D_a"] == 0,["D_a", "a", "a_squared"]])]
        )
plt.plot(a[cutoff:], 
         [x[0] for x in lr.predict(in_df.ix[in_df["D_a"] == 1, ["D_a", "a", "a_squared"]])]
        )

print("alpha:", lr.intercept_, "\ncoefs:", lr.coef_)
```

Second, we can subtract $a_0$ such that we allow the running variable coefficient to be different before and after the dummy is switched on;
$$\bar{M_a} = \alpha + \rho D_a + \gamma_1 (a - a_0) + \delta[(a - a_0)D_a] + e_a$$

**I don't get this.**  
Let's talk it through. So before the transition, the regression is like so:

$$\bar{M_a} = \alpha + \gamma_1 (a - a_0) + e_a$$
The $\gamma_1$ component will aways be negative and approaching zero. After the the transition, the regression is like this:

$$\bar{M_a} = \alpha + \rho + \gamma_1 (a - a_0) + \delta (a - a_0) + e_a$$

So we've added an extra $\rho + \delta (a - a_0)$ (the treatment effect) where the $\rho$ is the 'discontinuity' and $\delta$ is the change in slope of the post-$D_a$ line.  

**But why bother subtracting $a_0$?**  
This is so that $\rho$ is still the 'discontinuity'. Consider at the exact point the dummy is switched on (a - $a_0$ = 0):  
*Before*
$$\bar{M_a} = \alpha + e_a$$
*After*
$$\bar{M_a} = \alpha + \rho + e_a$$

This is what is already being done in the data in the book (i.e. having a_0 (the cutoff) be at a = 0). In many cases for regression discontiuity this is not necessary, but in the above case it is (for simple interpretation).

```python
#Dual regression case:

#Create data:
x_axis = np.arange(0, 200)
cutoff = round(len(x_axis) / 2)
y_axis = list(map(lambda x : int(abs(np.random.normal(40 + 10*(x>cutoff), 8)) - 0.01 * x), x_axis))

#Massage data into nice structure:
M_a = y_axis
D_a = x_axis >= cutoff
#(a - a_0):
a = x_axis - cutoff
#delta:
Delta = D_a * a

#Massage data into DF:
in_df = pd.DataFrame(data = {"D_a" : D_a, 
                             "a" : a, 
                             "Delta" : Delta,
                             "M_a" : M_a}, index = a)

#Run regression
lr = LinearRegression()
lr.fit(X = in_df[["D_a", "a", "Delta"]], y = in_df[["M_a"]])

plt.plot(x_axis - cutoff, y_axis)
plt.xlabel("Months Before/After Age 21")
plt.ylabel("Monthly Deaths")

#Function for getting result:
def reg_f(x1, x2, x3, lr):
    return(lr.intercept_[0] + x1 * lr.coef_[0][0] + x2 * lr.coef_[0][1] + x3 * lr.coef_[0][2])

#Plot
cutoff = 0
plt.plot([start, cutoff], [reg_f(0, a[0], Delta[0], lr), reg_f(0, cutoff, cutoff, lr)])
plt.plot([cutoff, end], [reg_f(1, cutoff, cutoff, lr), reg_f(1, a[-1], Delta[-1], lr)])
print("alpha:", lr.intercept_, "\ncoefs:", lr.coef_)
```

We can also combine these teqniques to get different non-linear effects before and after the cutoff.

### Non-parametric RD
Non-linearity becomes less of a problem as we narrow the band of data points we are examining one either side of $a_0$. The drawback here is that we have less points and thus less statistical confidence in our observations.

### Fuzzy RD
In fuzzy regression discontinuity, rather than having a hard cutoff we have to use a 2 step least squares regression (2SLS).

Reduced Form:  
$$Y_i = \alpha_2 + \lambda \bar{X}_{(i)} + \beta_2 R_i + e_{2i}$$

First Stage:
$$\bar{X}_{(i)} = \alpha_1 + \phi \bar{D}_{i} + \beta_1 R_i + e_{1i}$$
